#### Crear un tablero de ajedrez con 1 en las negras y 0 en las blancas

```python
a = np.zeros([8,8])
a.astype(int)
```

```python
for i in range(4):
    for j in range(8):
        if j%2 == 0:
            a[j, 2*i + 1]=1
        else:
            a[j, 2*i]=1
a
```
