```python
import numpy as np
def Pas(n):
    P = np.zeros((n,n))
    P[::, 0] = 1
    for a in range(n):
        P[a][a] = 1
    for i in range(2, n):
        for j in range(1, i):
            P[i, j] = P[i-1,j-1] + P[i-1, j]
    print(P)
```

```python
Pas(4)
```

```python
Pas(6)
```
