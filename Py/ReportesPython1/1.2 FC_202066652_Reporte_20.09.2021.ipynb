```python
def f(x):
    return np.exp(- x**2)
```

```python
x = np.linspace(-1, 5, 50)
```

```python
plt.plot(x, f(x), label="Funcion $f(x)=e^{-x^{2}}$")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.legend()
plt.title("Funcion $f(x)=e^{-x^{2}}$")
```

```python
x0 = np.linspace(-1, 5, 15)
plt.plot(x, f(x), 'ro')
plt.plot(x0, 1 - f(x0), 'go')
```

```python
x0 = np.linspace(-1, 5, 25)
plt.plot(x, f(x), 'ro')
plt.plot(x0, 1 - f(x0), 'g--')
```

```python
x0 = np.linspace(-1, 5, 25)
plt.plot(x, f(x), c='r', ls='', marker='*')
plt.plot(x0, 1 - f(x0), c='g', ls='', marker='*')
```

```python
def f(x, y):
    return np.cos(x) + np.sin(y) ** 2
```

```python
x = np.linspace(-2, 2)
y = np.linspace(-2, 2)
xx, yy = np.meshgrid(x, y)

plt.contour(xx, yy, f(xx, yy))
plt.colorbar()
```

```python
def f(x):
    return np.exp(-x ** 2)
x = np.linspace(-1, 5, 30)
f(x)
```

```python
fig, axes = plt.subplots()

axes.plot(x, f(x), 'ro', label="Funcion $f(x)=e^{-x^{2}}$")
axes.legend()
axes.set_xlim(-2, 4)
axes.set_ylim(-1, 2)
```
