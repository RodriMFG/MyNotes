```python
from IPython.display import Markdown, display, Math
from sympy import*
import sympy as sp
sp.init_printing()
import math
import numpy as np

def triang(n, a, b):
    xs = np.zeros(n)
    for i in range(n):
        s = 0
        for j in range(i):
            s += a[i,j]*xs[j]
        xs[i] = (b[i] - s)/a[i,i]
        display(Math(rf'\text{{Valor aproximado...}}\
        x_{{%s}} \cong %s' %(i+1,np.round(xs[i],4)) ))
    return xs
```
