### 1.2 Fuentes de error en la resolucion de estos problemas

Sea x la medida del lado de los cuadrados

```python
def f(v, p): #donde v es el volumen deseado y p la precision del calculo (entre mas bajo mas preciso)
    x = 0
    if v>1500: #1500 cc es aproximadamente el volumen maximo
        return 'Excede volumen maximo'
    while (32 - 2*x)*(24 - 2*x)*x<=v:
        x += p
        if (32 - 2*x)*(24 - 2*x)*x==v:
            x1 = x
        elif (32 - 2*x)*(24 - 2*x)*x>v:
            x1 = (x - p/2) #Una aproximacion del valor de x
    x = x1 + p/2
    while (32 - 2*x)*(24 - 2*x)*x>=v:
        x += p
        if (32 - 2*x)*(24 - 2*x)*x==v:
            x2 = x
        elif (32 - 2*x)*(24 - 2*x)*x<v:
            x2 = x - p/2
    print(round(x1,4),"cm", "o", round(x2,4),"cm")
```

```python
f(1000, 0.5)
```

```python
f(1000, 0.1)
```

```python
f(1000, 0.0001)
```

```python
f(100, 0.001)
```

```python
f(1500, 0.0001)
```

```python
f(1600, 1)
```
