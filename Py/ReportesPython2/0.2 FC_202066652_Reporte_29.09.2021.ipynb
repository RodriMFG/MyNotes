### Metodo de Heron

donde 'p' es el numero cuya raiz queremos aproximar

```python
from IPython.display import Markdown, display, Math

#p: numero
#x: valor incicial
#n iteraciones
def raizc(p, x, n):
    for i in range(n):
        x = (1/2)*(x + p/x)
        display(Math(rf'\text{{Para }} x_{{%s +1}} = \frac{1}{2}\
        (x_{{%s}} + %s/x_{{%s}}) = %s \
        \\ \Rightarrow x_{{%s +1}}^{2}=%s' %(i,i,p,i,x,i,x**2)))
        
```

```python
raizc(7, 3, 3)
```

```python
raizc(7, 3, 10)
```
