### Ejemplo de Metodo Iterativo

Calculo de raices de ordenes mayore o iguales a 2

Usando derivadas y una aproximacion inicial podemos aproximar de manera precisa las raices cuadradas, cubicas, etc.

Sea $a$ el radicando y $b$ el radical

Definimos $x=a^{\frac{1}{b}} \leftrightarrow x^{b}=a \leftrightarrow x^{b}-a=0$

Y $f(x)=x^{b}-a$

Derivando obtenemos $\frac{d}{dx}({x^{b}-a}{x})=\frac{d}{dx}(0) \leftrightarrow bx^{b-1}=0$

Elegimos una aproximacion $x_1=c$

Y tenemos la formula recursiva $x_{n+1}=x_{n}-\frac{f(x)}{f'(x)}=x_{n}-\frac{x^{b}-a}{bx^{b-1}}$

Hagamos una funcion en python que pueda aproximar estas raices:

```python
#r=radicando
#p=radical
#c=aprox inicial
#n=no. de iteraciones
def R(r, p, c, n):
    
    x=c
    for i in range(n):
        x = x - (x**p - r)/(p*x**(p-1))
    return x
```

```python
R(8, 3, 3, 6)
```

```python
R(2, 2, 1.5, 5)
```
